```python
import numpy as np
import pylab as pl
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline
mymean=100        #all the distributions have a mean=100
mysize= np.random.random_integers(11, 1999, 100)   #the sizes of the distributions are given by random numbers on the interval.
```

```python
#Creating a dictionary for my distributions:
dist = {} 
dist['chq'] = {} 
dist['norm'] = {}
dist['pois'] = {}
dist['bin'] = {}
dist['exp'] = {}
dist['gam'] = {}
#random numbers for each distribution
for n in mysize:
    dist['chq'][n] = np.random.chisquare(mymean,n) # Chisquare
    dist['norm'][n] = np.random.normal(mymean,1,n) # Normal/Gaussian with sd=1
    dist['pois'][n] = np.random.poisson(mymean,n) # Poisson
    dist['bin'][n] = np.random.binomial(2*mymean,0.5,n) # Binomial
    dist['exp'][n] = np.random.exponential(mymean,n)  #Exponential distribution
    dist['gam'][n] = np.random.gamma(mymean, scale=1.0, size=n)  #gamma distribution
```

```python
means = {} # Dictionary for means
means['chq'] = {}
means['norm'] = {}
means['pois'] = {}
means['bin'] = {}
means['exp'] = {}
means['gam'] = {}
stds= {} #Dictionary for standard deviations
stds['chq'] = {}
stds['norm'] = {}
stds['pois'] = {}
stds['bin'] = {}
stds['exp'] = {}
stds['gam'] = {}
fig = pl.figure(figsize=(20,60))

count=0
for D in dist:
    title= 'Mean for %s'%(D)
    count=count+1
    for i in dist[D].iterkeys():
        ax_mu_n = fig.add_subplot(6,2,count)
        means[D][i] = dist[D][i].mean()
        #and plot it
        ax_mu_n.plot(i, means[D][i], 'o')
        ax_mu_n.set_xlabel('sample size')
        ax_mu_n.plot([min(mysize), max(mysize)],[mymean,mymean], 'k')
        ax_mu_n.set_title(title)
plt.show()
fig = pl.figure(figsize=(20,60))
count=0
for D in dist:
    title= 'Standard Deviation for %s' %(D)
    count=count+1
    for i in dist[D].iterkeys():
        ax_mu_n = fig.add_subplot(6,2,count)
        stds[D][i]=np.std(dist[D][i])
        #and plot it
        ax_mu_n.plot(i, stds[D][i], 'o')
        ax_mu_n.set_xlabel('sample size')
        ax_mu_n.plot([min(mysize), max(mysize)],[10,10], 'k')
        ax_mu_n.set_title(title)
plt.show()





```

In the plot for the mean we can see that the different sample's means tend to be grouped around the line of the given mean for the generation of values (in this case, mymean=100). In the case of smaller sample size, this value differ from the given mean, but the whole set of means converge to the given mean. 

In the case of the standard deviation, it happens the same: the different standard deviations tend to be grouped around the line of the given standard deviation for the generation of values. Again, in the case of smaller sample size, this value differ from the given standard deviation, but the whole set of standard deviations converge to the given sd. 

This behavior is expected, because we are using functions for random values generation that follow given distributions with a fixed value for the mean and the standard deviations.
